```python
from sklearn import tree
from sklearn import preprocessing
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv(r"C:\Users\Administrator\Desktop\sunrise Hackathon\ICRISAT Data.csv")
```

```python
labels=data['Crop before']
print(labels)
```

```python
data=data.drop('Crop before',axis=1)
```

```python
print(data)
```

```python
districts = preprocessing.LabelEncoder()
districts.fit(list(data['District']))
```

```python
districts.transform(["Anantapur"])
```

```python
mandal = preprocessing.LabelEncoder()
mandal.fit(list(data['Mandal']))
```

```python
mandal.transform(['Penukonda'])
```

```python
soil = preprocessing.LabelEncoder()
soil.fit(list(data['Soil type']))
```

```python
village = preprocessing.LabelEncoder()
village.fit(list(data['Village']))
```

```python
features=[]

for i in range(4560):
    features.append(data.iloc[i])
```

```python
print(len(features))
```

```python
for f in features:
    f[0]=districts.transform([f[0]])
    f[1]=mandal.transform([f[1]])
    f[2]=village.transform([f[2]])
    f[3]=soil.transform([f[3]])
```

```python
from sklearn.model_selection import train_test_split
```

```python
feature_train,feature_test,label_train,label_test = train_test_split(features,labels,test_size=0.2)
```

```python
print(len(feature_train),len(label_train))
```

```python
print(len(feature_test),len(label_test))
```

```python
clf = tree.DecisionTreeClassifier()
```

```python
clf = clf.fit(feature_train, label_train)
```

```python
l=[]
for feature in feature_test:
    l.append(clf.predict([feature]))
```

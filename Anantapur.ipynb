```python
from sklearn import tree
from sklearn import preprocessing
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv(r"C:\Users\Administrator\Desktop\sunrise Hackathon\Anantapur.csv")
```

```python
labels=data['Crop before']
print(labels)
```

```python
data=data.drop('Crop before',axis=1)
```

```python
print(data)
```

```python
features=data.columns.values
print(features)
```

```python
districts = preprocessing.LabelEncoder()
districts.fit(list(data['District']))
```

```python
districts.transform(["Anantapur"])
```

```python
mandal = preprocessing.LabelEncoder()
mandal.fit(list(data['Mandal']))
```

```python
mandal.transform(['Raptadu'])
```

```python
soil = preprocessing.LabelEncoder()
soil.fit(list(data['Soil type']))
```

```python
village = preprocessing.LabelEncoder()
village.fit(list(data['Village']))
```

```python
features=[]

for i in range(140):
    features.append(data.iloc[i])
    
```

```python
features[0][0]
```

```python
for f in features:
    f[0]=districts.transform([f[0]])
    f[1]=mandal.transform([f[1]])
    f[2]=village.transform([f[2]])
    f[3]=soil.transform([f[3]])
```

```python
print(features)
```

```python
clf = tree.DecisionTreeClassifier()
```

```python
clf = clf.fit(features, labels)
```

```python
print( clf.predict([[0,0,2,2,6.19,0.07,7.13,41,587,101,5.16,0.3,0.17,8.89,0.51,15.24]]))
```
